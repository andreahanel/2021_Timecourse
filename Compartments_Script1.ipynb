```r
location_GO <- str_extract_all (d$GO_cc, regex ('Endoplasmic reticulum|Golgi apparatus|Lysosome|plasma membrane|Mitochondrion|Nucleus|Peroxisome|Cytoplasm|Cytoskeleton|extracellular region|Endosome|Cytoplasmic vesicle',ignore_case = TRUE)
                                               )
```

```r
head (location_GO)
```

```r
compartment_GO <- (location_GO)  %>% 
map_chr(~ str_c(unique(na.omit(.x)), collapse = ", ")) # %>%
compartment_GO  <- as.data.frame(compartment_GO)
head (compartment_GO)
```

```r
compartment_GO$compartment_GO <- sub("endoplasmic reticulum","endoplasmic reticulum",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("golgi apparatus","Golgi apparatus",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("lysosome","lysosome",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("plasma membrane","cell membrane",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("mitochondrion","mitochondrion",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("nucleus","nucleus",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("peroxisome","peroxisome",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("cytoplasm","cytoplasm",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("cytoskeleton","cytoskeleton",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("extracellular region","secreted",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("endosome","endosome",as.character(compartment_GO$compartment_GO))
compartment_GO$compartment_GO <- sub("cytoplasmic vesicle","cytoplasmic vesicle",as.character(compartment_GO$compartment_GO))
```

```r
location <- str_extract_all (d$Keywords, regex('Endoplasmic reticulum|Golgi apparatus|Lysosome|Cell membrane|Mitochondrion|Nucleus|Peroxisome|Cytoplasm|Cytoskeleton|Secreted|Endosome|Cytoplasmic vesicle')
                            )
```

```r
compartment <- (location)  %>% 
map_chr(~ str_c(unique(na.omit(.x)), collapse = ", ")) # %>%
#str_to_lower() 
compartment  <- as.data.frame(compartment)
compartment # %>% 
#str_to_lower()
```

```r
compartment$compartment <- tolower (compartment$compartment)
compartment
```

```r
fct_short <- cbind(genes,compartment, compartment_GO)
```

```r
jarod = fct_short %>%
column_to_rownames ("genes")
```

```r
j2 <- jarod %>%
filter (., compartment == "")

jnames <- rownames (j2)
dim  (j2)
```

```r
d <- read.table ("./data/f_time2020_inlog_FC15_FC2_evidence-threshold.tsv", sep = '\t', header = T, row.names =1)

d = d %>%
rownames_to_column ("genes")

d = d %>%
select (c(genes, description, howmany))  #row.names =1
```

```r
head (d)
dim (d)
```

```r
mrg = merge(x=d,y=fct_short,by.x="genes", all.x =T)
```

```r
head (mrg)
dim (mrg)
```

```r
mrg_sisa = mrg %>%
mutate(.,
       compartment_final = case_when(
         compartment == "" ~ compartment_GO,
       TRUE ~ compartment))
     
mrg_sisa$compartment_final = gsub("golgi apparatus","Golgi apparatus",as.character(mrg_sisa$compartment_final))
```

```r
mrg_sisa = mrg_sisa %>%
select (., genes, description, howmany, compartment_final)
```
