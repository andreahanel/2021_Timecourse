## Modelmatrix

```r
design <- model.matrix(~0+Group) 
colnames(design) <- levels(Group)
head (design) 
```

```r
#Estimate dispersion
y <- estimateDisp(y, design, robust=TRUE) #The estimation is also robustiﬁed against potential outlier genes.
y$common.dispersion
sqrt(y$common.dispersion)

summary (sqrt(y$trended.dispersion))
summary(sqrt(y$tagwise.dispersion))

BCV = plotBCV(y, cex = 0.1)
```

```r
fit <- glmQLFit(y,design, robust=TRUE) #"Setting robust=TRUE in glmQLFit is strongly recommended" (edgeRUsersGuide)
plotQLDisp(fit)
summary(fit$df.prior)
```

```r
#trt <- (sampleinfo[grepl("^V", sampleinfo$treatment),])
trtU <- unique (sampleinfo$Group)
trtU
```

```r
my.contrasts <- makeContrasts(
    
    V125D3_04h_EtOH_04h = 
    V125D3_04h - EtOH_04h, 
    
    V125D3_08h_EtOH_08h = 
    V125D3_08h - EtOH_08h, 
    
    V125D3_24h_EtOH_24h = 
    V125D3_24h - EtOH_24h, 

    V125D3_48h_EtOH_48h = 
    V125D3_48h - EtOH_48h,
    
    levels = colnames(design))
```

```r
tr_1 <- glmTreat(fit, contrast=my.contrasts[,"V125D3_04h_EtOH_04h"],lfc=log2(1.5)) 
tr_2 <- glmTreat(fit, contrast=my.contrasts[,"V125D3_08h_EtOH_08h"],lfc=log2(1.5)) 
tr_3 <- glmTreat(fit, contrast=my.contrasts[,"V125D3_24h_EtOH_24h"],lfc=log2(2)) 
tr_4 <- glmTreat(fit, contrast=my.contrasts[,"V125D3_48h_EtOH_48h"],lfc=log2(2)) 
```

```r
sumTr_1 <- summary(decideTests(tr_1, method="separate", adjust.method="BH", p.value=0.05,lfc=0)) 
sumTr_2 <- summary(decideTests(tr_2, method="separate", adjust.method="BH", p.value=0.05,lfc=0)) 
sumTr_3 <- summary(decideTests(tr_3, method="separate", adjust.method="BH", p.value=0.05,lfc=0)) 
sumTr_4 <- summary(decideTests(tr_4, method="separate", adjust.method="BH", p.value=0.05,lfc=0)) 
```

```r
sumTr005 <- cbind (sumTr_1, 
                   sumTr_2, 
                   sumTr_3, 
                   sumTr_4
                  ) %>% # 
  t()

sumTr005
```

```r
top_04h_long <- 
topTags (
    tr_1, n=Inf, sort.by = "none",  adjust.method ="BH", 
                       #  p.value=0.05
                              )
top_08h_long <- 
topTags (
    tr_2, n=Inf, sort.by = "none",  adjust.method ="BH", 
                       #  p.value=0.05
                              )
top_24h_long <- 
topTags (
    tr_3, n=Inf, sort.by = "none",  adjust.method ="BH", 
                       #  p.value=0.05
                              )
top_48h_long <- 

topTags (
    tr_4, n=Inf, sort.by = "none",  adjust.method ="BH", 
                       #  p.value=0.05
    )
```

```r
head (top_04h_long)
```

```r
top_04h_long <- 

as.data.frame(top_04h_long) %>%
mutate(absLogFC_04h = abs(logFC)) %>%
select(., -(PValue))%>%
  rename(., 
         logFC_04h = logFC, #shrunk log2-fold-change of expression between conditions being tested.
        # unshrunk.log2FC_04h = unshrunk.logFC, #unshrunk log2-fold-change of expression between conditions being tested. Exists only when prior.count is not equal to 0 for glmfit
         logCPM_04h = logCPM, #average log2-counts per million, the average taken over all libraries.
         FDR_04h = FDR) 


top_08h_long <- 

as.data.frame(top_08h_long) %>%
  mutate(absLogFC_08h = abs(logFC)) %>%
select(., -(PValue))%>%
  rename(., 
         logFC_08h = logFC, 
         logCPM_08h = logCPM,
         FDR_08h = FDR) 


top_24h_long <- 

as.data.frame(top_24h_long) %>%
mutate(absLogFC_24h = abs(logFC)) %>%
select(., -(PValue))%>%
  rename(., 
         logFC_24h = logFC, 
         logCPM_24h = logCPM,
         FDR_24h = FDR) 

top_48h_long <- 

as.data.frame(top_48h_long) %>%
  mutate(absLogFC_48h = abs(logFC)) %>%
select(., -(PValue))%>%
  rename(., 
         logFC_48h = logFC, 
         logCPM_48h = logCPM,
         FDR_48h = FDR)
```

```r
z4 <- z3 %>%
  rownames_to_column("genes") %>%

mutate(EtOH_04h.Avg = rowMeans(select(., matches("EtOH_04h"))) ) %>%
  mutate(V125D3_04h.Avg = rowMeans(select(., matches("V125D3_04h"))) ) %>% 
  
mutate(EtOH_08h.Avg = rowMeans(select(., matches("EtOH_08h"))) ) %>% 
  mutate(V125D3_08h.Avg = rowMeans(select(., matches("V125D3_08h"))) ) %>% 
  
mutate(EtOH_24h.Avg = rowMeans(select(., matches("EtOH_24h"))) ) %>% 
  mutate(V125D3_24h.Avg = rowMeans(select(., matches("V125D3_24h"))) ) %>% 

mutate(EtOH_48h.Avg = rowMeans(select(., matches("EtOH_48h"))) ) %>% 
  mutate(V125D3_48h.Avg = rowMeans(select(., matches("V125D3_48h"))) )  

str (z4)
```

```r
z4$sum_avg_expr = (z4$EtOH_04h.Avg+z4$EtOH_08h.Avg+z4$EtOH_24h.Avg+z4$EtOH_48h.Avg+ z4$V125D3_04h.Avg+z4$V125D3_08h.Avg+z4$V125D3_24h.Avg+z4$V125D3_48h.Avg)
head (z4)
```

```r
z4$evidence_threshold = 
  case_when(
  z4$sum_avg_expr < 25 ~ "no",
  z4$sum_avg_expr >= 25 ~ "yes"
  )
```
